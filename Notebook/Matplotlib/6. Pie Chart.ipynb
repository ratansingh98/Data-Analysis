```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values = [1400,600,300,410,250]
exp_labels = ["Home Rent",'Food',"Phone/Internet Bill","Car",'Other Utilities']
```

```python
plt.axis("equal")
plt.pie(exp_values,labels = exp_labels, radius=1.5,autopct= '%.1f%%', shadow= True, )
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels = exp_labels, radius=1.5,autopct= '%.1f%%', shadow= True, explode =[0,0,0,0.2,0.2],
        startangle=180 )
plt.show()
```
