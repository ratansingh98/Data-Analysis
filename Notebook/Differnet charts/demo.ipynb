```python
df1 = pd.DataFrame(np.random.randn(20,4),columns= list("abcd"))
temp = pd.crosstab(df['b'].head(2),df1['b'].head(2))
temp.plot(kind = 'bar', stacked=True, color =['red','blue'],grid = False)
```

```python
fig=plt.figure()

#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)

#Variable
ax.hist(df['a'],bins = 5)

#Labels and Tit
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('#Employee')
plt.show()
```

```python
from pandas.plotting import table

# sample data
raw_data = {'jobs': ['Study', 'Coding', 'Sleep', 'Music', 'Fun'],
        'jan': [4, 24, 31, 2, 3],
        'feb': [25, 94, 57, 62, 70],
        'march': [5, 43, 23, 23, 51]}
df = pd.DataFrame(raw_data, columns = ['jobs', 'jan', 'feb', 'march'])
df['total'] = df['jan'] + df['feb'] + df['march']
plt.figure(figsize=(26,10))
# plot chart
ax1 = plt.subplot(131, aspect='equal')
df.plot(kind='pie', y = 'total', ax=ax1, autopct='%1.2f%%', 
 startangle=0, shadow=False, labels=df['jobs'], legend = True, fontsize=18)
```
